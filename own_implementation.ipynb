```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.animation as animation
from matplotlib.animation import PillowWriter
```

# Advection equation sample code via difference method

```python
fig = plt.figure()

## Conditions
# Before use Triangular initial condition such as
x_init = 10
x_sep = 10
fx_init = 0.5

c = 0.5

ims = []

# linespaces condition
x_line = 100
x_step = 201
x = np.linspace(0, x_line, x_step)

# separate times
dt = 0.5

# separate spaces and initial f condition
dx = x[ 2] - x[ 1]
f = [0] * x_step
fn = [0] * x_step

# Then I use Triangular initial condition as like
for i in range(0, x_step):
    if (x[i] >= (x_init - x_sep) and x[i] <= x_init):
        f[i] = fx_init/x_sep*(x[i] - x_init + x_sep)
    elif (x[i] > x_init and x[i] < (x_init+x_sep)):
        f[i] = fx_init/x_sep*(x_init + x_sep - x[i])
    else:
        f[i]=0

# Plots settings
im = plt.title("Sample of advection equation")
im = plt.xlabel("x")
im = plt.ylabel("f")

# Load time
start_time = 0
end_time = 200
while (start_time <= end_time):
    for i in range(1, x_step-1):
        fn[i] = f[i] - c*(f[i] - f[i-1])*dt/dx

    f = fn

    # Plots
    im = plt.plot(x,f,'r')
    ims.append(im)
    start_time = dt + start_time

# Animations
ani = animation.ArtistAnimation(fig, ims, interval=10)
plt.show();
ani.save('results/advec_results.gif',writer='imagemagick');
```

# Diffusion equation sample code via difference method

```python
fig = plt.figure()

## Conditions
x_init = 50
x_sep = 10
fx_init = 0.5

D = 0.1  # 拡散係数

ims = []

# linespaces condition
x_line = 100
x_step = 201
x = np.linspace(0, x_line, x_step)

# separate times
dt = 1

# separate spaces and initial f condition
dx = x[2] - x[1]
f = np.zeros(x_step)
fn = np.zeros(x_step)

# Then I use Triangular initial condition as like
for i in range(0, x_step):
    if (x[i] >= (x_init - x_sep) and x[i] <= x_init):
        f[i] = fx_init / x_sep * (x[i] - x_init + x_sep)
    elif (x[i] > x_init and x[i] < (x_init + x_sep)):
        f[i] = fx_init / x_sep * (x_init + x_sep - x[i])
    else:
        f[i] = 0

# Plots settings
plt.title("Sample of diffusion equation")
plt.xlabel("x")
plt.ylabel("f")

# Load time
start_time = 0
end_time = 200
while (start_time <= end_time):
    for i in range(1, x_step - 1):
        fn[i] = f[i] + D * (f[i + 1] - 2 * f[i] + f[i - 1]) * dt / (dx ** 2)

    f = fn.copy()

    # Plots
    im = plt.plot(x, f, 'r')
    ims.append(im)
    start_time = dt + start_time

# Animations
ani = animation.ArtistAnimation(fig, ims, interval=10)
plt.show()
ani.save('results/diff_results.gif', writer='imagemagick')
```
